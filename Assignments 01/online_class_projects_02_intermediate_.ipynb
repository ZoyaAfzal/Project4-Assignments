```python
# üåç Planetary Weight Calculator

def planetary_weight_calculator():
    print('üåç Welcome to the Planetary Weight Calculator!')
    print('---------------------------------------------')

    try:
        earth_weight = float(input('‚öñÔ∏è  Enter your weight on Earth (kg): '))
    except ValueError:
        print("‚ùå Please enter a valid number for your weight.")
        return

    gravity_ratios = {
        "Mercury": 0.38,
        "Venus": 0.91,
        "Mars": 0.38,
        "Jupiter": 2.34,
        "Saturn": 1.06,
        "Uranus": 0.92,
        "Neptune": 1.19
    }

    print('\nü™ê Select a planet to see what you‚Äôd weigh there:')
    for planet in gravity_ratios:
        print(f'  - {planet}')

    planet_choice = input('\n Enter the name of the planet: ').title()

    if planet_choice in gravity_ratios:
        new_weight = earth_weight * gravity_ratios[planet_choice]
        print(f'\n On {planet_choice}, you would weigh approximately: {new_weight:.2f} kg')
    else:
        print(f" Sorry, '{planet_choice}' is not in our planetary list.")

    print('\n Thank you for using the calculator. Safe space travels! üöÄ')


if __name__ == '__main__':
    planetary_weight_calculator()
```

```python
# Lists and Dicts

print('üçá Problem #2: Index Game üçä')

def access_elements(fruit_list, index):
    """Returns the element at the given index in the list if it exists, otherwise returns Index out of range!"""
    if 0 <= index < len(fruit_list):
        return f'Element at index {index}: {fruit_list[index]}'
    return 'Index out of range!'

def modify_element(fruit_list, index, new_val):
    """Modifies the element at the given index in the list to the new value if exists, otherwise returns Index out of range!"""
    if 0 <= index < len(fruit_list):
        old_val = fruit_list[index]
        fruit_list[index] = new_val
        return f'Element "{old_val}" at index {index} modified to "{new_val}"'
    return 'Index out of range!'

def slice_list(fruit_list, start, end):
    """Returns a new list containing the elements from the start index to the end index!"""
    if 0 <= start < len(fruit_list) and 0 <= end <= len(fruit_list) and start < end:
        return f'Sliced list from index {start} to {end}: {fruit_list[start:end]}'
    return 'Invalid start or end index!'

def list_game():
    print('\nWelcome to the List Manipulation Game!')
    print('Fruits List Game üçäüçáüçç')

    fruit_list = ['apple', 'banana', 'orange', 'grape', 'pineapple']

    while True:
        print('\n Current List:', fruit_list)

        print('\nWhat would you like to do?')
        print('1Ô∏è‚É£ Access an Element')
        print('2Ô∏è‚É£ Modify an Element')
        print('3Ô∏è‚É£ Slice the List')
        print('4Ô∏è‚É£ Quit')

        choice = input('üëâ Enter your choice (1-4): ').strip()

        if choice == '1':
            try:
                index = int(input('Enter the index you want to access: '))
                print(access_elements(fruit_list, index))
            except ValueError:
                print('‚ö†Ô∏è Please enter a valid number.')
        elif choice == '2':
            try:
                index = int(input('Enter the index you want to modify: '))
                new_val = input('Enter the new value: ')
                print(modify_element(fruit_list, index, new_val))
            except ValueError:
                print('‚ö†Ô∏è Please enter a valid number.')
        elif choice == '3':
            try:
                start = int(input('Enter the start index: '))
                end = int(input('Enter the end index: '))
                print(slice_list(fruit_list, start, end))
            except ValueError:
                print('‚ö†Ô∏è Please enter valid numbers.')
        elif choice == '4':
            print('\nüëã Exiting the game. Thanks for playing! üçâ See you again!')
            break
        else:
            print('üö´ Invalid choice. Please choose between 1 and 4.')

if __name__ == '__main__':
    list_game()

```
