```python
import random

# Word categories and their words
categories = {
    'python': ['function', 'variable', 'loop', 'class', 'decorator', 'lambda', 'module', 'list', 'tuple'],
    'typescript': ['interface', 'any', 'function', 'typescript', 'promise', 'type', 'decorator', 'async'],
    'javascript': ['callback', 'promise', 'closure', 'event', 'array', 'function', 'scope', 'hoisting', 'object']
}

def choose_category():
    print("Choose a category:")
    print("1. Python")
    print("2. TypeScript")
    print("3. JavaScript")

    choice = input("Enter the number corresponding to your choice: ")
    if choice == '1':
        return 'python'
    elif choice == '2':
        return 'typescript'
    elif choice == '3':
        return 'javascript'
    else:
        print("Invalid choice. Defaulting to Python.")
        return 'python'

def choose_difficulty():
    print("\nChoose a difficulty level:")
    print("1. Easy (8 attempts)")
    print("2. Medium (6 attempts)")
    print("3. Hard (4 attempts)")

    choice = input("Enter the number corresponding to your choice: ")
    if choice == '1':
        return 8  # Easy
    elif choice == '2':
        return 6  # Medium
    elif choice == '3':
        return 4  # Hard
    else:
        print("Invalid choice. Defaulting to Medium.")
        return 6  # Default to Medium

def get_word(category):
    return random.choice(categories[category])

def display_word(word, guessed_letters):
    display = ""
    for letter in word:
        if letter in guessed_letters:
            display += letter + " "
        else:
            display += "_ "
    return display.strip()

def hangman():
    print("Welcome to Hangman!")

    # Choose category and difficulty
    category = choose_category()
    word = get_word(category)
    max_attempts = choose_difficulty()

    guessed_letters = []
    attempts_left = max_attempts
    word_guessed = False

    print(f"\nCategory: {category.capitalize()}")
    print(f"The word has {len(word)} letters.\n")

    while attempts_left > 0 and not word_guessed:
        print(f"Attempts left: {attempts_left}")
        print(display_word(word, guessed_letters))

        guess = input("Guess a letter: ").lower()

        if guess in guessed_letters:
            print("You already guessed that letter.")
        elif len(guess) != 1 or not guess.isalpha():
            print("Please enter a single letter.")
        else:
            guessed_letters.append(guess)

            if guess not in word:
                attempts_left -= 1
                print(f"Sorry, {guess} is not in the word.")
            else:
                print(f"Good job! {guess} is in the word.")

        # Check if the word is fully guessed
        if all(letter in guessed_letters for letter in word):
            word_guessed = True

    if word_guessed:
        print(f"\nCongratulations! You guessed the word: {word}")
        return True
    else:
        print(f"\nYou lost! The word was: {word}")
        return False

def play_game():
    wins = 0
    losses = 0

    while True:
        # Play the hangman game
        if hangman():
            wins += 1
        else:
            losses += 1

        # Show current score
        print(f"\nScore: Wins: {wins} | Losses: {losses}")

        # Ask if the user wants to play again
        play_again = input("\nDo you want to play again? (yes/no): ").lower()
        if play_again != 'yes':
            print(f"\nThanks for playing! Final Score: Wins: {wins} | Losses: {losses}")
            break

# Start the game
play_game()
```
