```python
print('00_count_nums')

def count_numbers():
    count_dict = {}

    while True:
        user_input = input('Enter a number or type Exit to quit: ')
        if user_input.title() == 'Exit':
            break
        if user_input.isdigit():
            num = int(user_input)
            count_dict[num] = count_dict.get(num, 0) + 1
        else:
            print('Invalid input. Please enter a valid number or type Exit to quit.')

    return count_dict

def display_counts(count_dict):
    print('\nNumber Counts:')
    for num, count in count_dict.items():
        print(f'{num} appears {count} times.')

if __name__ == '__main__':
    counts = count_numbers()
    display_counts(counts)
```

```python
print('01_phonebook')

def add_contacts():
    phonebook = {}
    print("\nğŸ“˜ Add your contacts (leave name empty to finish):")
    while True:
        name = input("ğŸ‘¤Enter Name: ").strip()
        if name == "":
            break

        if name in phonebook:
            print(f"âš ï¸ '{name}' already exists with number {phonebook[name]}.")
            choice = input("Do you want to update the number? (y/n): ").strip().lower()
            if choice != 'y':
                continue

        number = input(f"ğŸ“ Phone number for {name}: ").strip()
        if not number.isdigit():
            print("âŒ Invalid number. Please enter digits only.\n")
            continue

        phonebook[name] = number
        print(f"âœ… Saved: {name} â†’ {number}\n")

    return phonebook

def show_contacts(phonebook):
    print("\nğŸ“’ Your Saved Contacts:")
    if not phonebook:
        print("No contacts found.")
        return
    for name, number in phonebook.items():
        print(f" - {name}: {number}")

def search_contact(phonebook):
    print("\nğŸ” Search your contacts (leave blank to go back):")
    while True:
        name = input("Name to search: ").strip()
        if name == "":
            break
        number = phonebook.get(name)
        if number:
            print(f"ğŸ“ {name}'s number: {number}")
        else:
            print(f"âŒ No contact found with the name '{name}'.")

def delete_contact(phonebook):
    print("\nğŸ—‘ï¸ Delete a contact (leave blank to go back):")
    while True:
        name = input("Name to delete: ").strip()
        if name == "":
            break
        if name in phonebook:
            confirm = input(f"Are you sure you want to delete '{name}'? (y/n): ").strip().lower()
            if confirm == 'y':
                del phonebook[name]
                print(f"âœ… '{name}' has been deleted.")
            else:
                print("âŒ Delete cancelled.")
        else:
            print(f"âŒ No contact found with the name '{name}'.")

def main():
    print("ğŸ“± Welcome to the Simple Phonebook App!")
    phonebook = add_contacts()

    while True:
        print("\nğŸ“‹ What would you like to do next?")
        print("1. Show all contacts")
        print("2. Search for a contact")
        print("3. Delete a contact")
        print("4. Exit")
        choice = input("Enter choice (1/2/3/4): ").strip()

        if choice == '1':
            show_contacts(phonebook)
        elif choice == '2':
            search_contact(phonebook)
        elif choice == '3':
            delete_contact(phonebook)
        elif choice == '4':
            print("\nğŸ‘‹ Goodbye! Have a great day.")
            break
        else:
            print("âŒ Invalid option. Please enter 1, 2, 3, or 4.")

if __name__ == '__main__':
    main()
```

```python
import hashlib

print('03_powerful_passwords')

def hash_password(password):
  return hashlib.sha256(password.encode()).hexdigest()

stored_logins = {
    'user@example.com': hash_password('password123'),
    'admin@example.com': hash_password('adminpass')
}

def login(email, password):
  hashed_password = hash_password(password)
  if email in stored_logins and stored_logins[email] == hashed_password:
    return True
  return False

if __name__ == '__main__':
  email = input('Enter your email: ')
  password = input('Enter your password: ')

  if login(email, password):
    print('Login successful')
  else:
    print('Login failed')
```
